**Data Preprocessing**

```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import SimpleImputer

imputer = SimpleImputer()
Preprocess_data=imputer.fit_transform(X)
Preprocesse_data=pd.DataFrame(Preprocess_data)
Preprocesse_data


data=Preprocesse_data
def cap_outliers_with_zscore(column, threshold=3):
    z_scores = np.abs((column - np.mean(column)) / np.std(column))
    mean = np.mean(column)
    std_dev = np.std(column)

    capped_column = np.where(z_scores > threshold, mean + threshold * std_dev, column)

    return capped_column

# Apply Z-score capping to each column in the DataFrame
for column in data.columns:
    data[column] = cap_outliers_with_zscore(data[column])

# print(data)
ppreprocesseed_data = pd.DataFrame(data)



from sklearn.preprocessing import QuantileTransformer
qt = QuantileTransformer(n_quantiles=10, random_state=0)
Preprocesses_data= qt.fit_transform(ppreprocesseed_data)

Preprocessed_dattaaa=pd.DataFrame(Preprocesses_data)
Preprocessed_dattaaa

```

**Feature Engineering**

```python
from sklearn.feature_selection import f_classif
from sklearn.feature_selection import SelectKBest
from sklearn.decomposition import PCA


Preprocessedd_data=SelectKBest(f_classif, k=700).fit_transform(Preprocessed_dattaaa, Y1)
Pppreprocesse_dataa=pd.DataFrame(Preprocessedd_data)


# Create a PCA instance with 500 components
pca = PCA(n_components=500)
# Fit PCA to your data
pca.fit(Pppreprocesse_dataa)
X_reduced = pca.transform(Pppreprocesse_dataa)

X_reduced=pd.DataFrame(X_reduced)
X_reduced
```

**Class Balancing**

```python

from imblearn.over_sampling import SMOTENC
smotenc = SMOTENC([1], random_state = 42)
X_SMOTE, Y_SMOTE = smotenc.fit_resample(X_reduced, Y1)


# from imblearn.combine import SMOTEENN
# # Create an instance of SMOTE-ENN
# smoteenn = SMOTEENN()
# # Fit and resample your dataset using SMOTE-ENN
# X_SMOTE, Y_SMOTE = smoteenn.fit_resample(X_reduced, Y1)

```

**Compare with other Boosting Classifiers**

```python
# Import necessary libraries

from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score
import lightgbm as lgb

# Assuming you already have X_train, X_test, Y_train, and Y_test

# Train and evaluate AdaBoost
ada_boost = AdaBoostClassifier(n_estimators=50)
ada_boost.fit(X_train, Y_train)
ada_boost_pred = ada_boost.predict(X_test)
ada_boost_accuracy = accuracy_score(Y_test, ada_boost_pred)
print("AdaBoost Accuracy:", ada_boost_accuracy)

# Train and evaluate Gradient Boosting (Scikit-learn)
gradient_boost = GradientBoostingClassifier(n_estimators=50)
gradient_boost.fit(X_train, Y_train)
gradient_boost_pred = gradient_boost.predict(X_test)
gradient_boost_accuracy = accuracy_score(Y_test, gradient_boost_pred)
print("Gradient Boosting Accuracy:", gradient_boost_accuracy)

# Train and evaluate LightGBM with histogram boosting
lgbm = LGBMClassifier(boosting_type='hist', n_estimators=50)
lgbm.fit(X_train, Y_train)
lgbm_pred = lgbm.predict(X_test)
lgbm_accuracy = accuracy_score(Y_test, lgbm_pred)
print("LightGBM (Histogram Boosting) Accuracy:", lgbm_accuracy)

# Train and evaluate Histogram Gradient Boosting

HGB_clf = HistGradientBoostingClassifier().fit(X_train, Y_train)
HGB_clf_pred = HGB_clf.predict(X_test)
hgb_hist_accuracy = accuracy_score(Y_test, HGB_clf_pred)
print("Histogram Gradient Boosting Accuracy:", hgb_hist_accuracy)
```

**Tuning the Model**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.model_selection import cross_val_score
from hyperopt import fmin, tpe, hp, Trials

# Load the breast cancer dataset
# dataset = load_breast_cancer()
X = X_train
y = Y_train

# Define the objective function
def objective(params):
    clf = HistGradientBoostingClassifier(**params)
    cv_scores = cross_val_score(clf, X, y, cv=5, scoring='accuracy')
    return 1 - np.mean(cv_scores)

# Define the hyperparameter space
space = {
    'learning_rate': hp.loguniform('learning_rate', 0, 0.8),
    'max_depth': hp.choice('max_depth', range(10, 28)),
    'min_samples_leaf': hp.choice('min_samples_leaf', range(1, 20)),
    'l2_regularization': hp.uniform('l2_regularization', 0, 0.5),
    'max_bins': hp.choice('max_bins', range(13, 49)),
    'warm_start': hp.choice('warm_start', [False, True]),
    'max_iter': hp.choice('max_iter', range(500, 1001))
}

# Run the optimization using Bayesian optimization with the Tree-structured Parzen Estimator (TPE) algorithm
trials = Trials()
best = fmin(objective, space, algo=tpe.suggest, max_evals=15, trials=trials)

# Print the best hyperparameters and their corresponding loss (accuracy)
print("Best hyperparameters:", best)
best_params = {
    'learning_rate': best['learning_rate'],
    'max_depth': best['max_depth'] + 3,
    'min_samples_leaf': best['min_samples_leaf'] + 1,
    'l2_regularization': best['l2_regularization'],
    'max_bins': best['max_bins'] + 32,
    'warm_start': best['warm_start'],
    'max_iter': best['max_iter']
}
clf = HistGradientBoostingClassifier(**best_params)

# Visualize accuracy vs. number of cross-validation folds
num_folds = list(range(2, 12))
accuracies = []
for fold in num_folds:
    cv_scores = cross_val_score(clf, X, y, cv=fold, scoring='accuracy')
    accuracies.append(np.mean(cv_scores))

plt.plot(num_folds, accuracies, marker='o')
plt.xlabel('Number of Cross-Validation Folds')
plt.ylabel('Accuracy')
plt.title('Accuracy vs. Number of Cross-Validation Folds')
plt.grid(True)
plt.show()
```
